```python
import os
import argparse

def analyze_video_files(directory):
    success_count = 0
    failure_count = 0

    # Walk through the directory and all subdirectories
    for root, _, files in os.walk(directory):
        for file in files:
            # Check if file is an .mp4
            if file.lower().endswith(".mp4"):
                filename_lower = file.lower()
                
                # Check for "success" or "failure" in the filename
                if "success" in filename_lower:
                    success_count += 1
                elif "failure" in filename_lower:
                    failure_count += 1

    total_runs = success_count + failure_count

    # Avoid division by zero if no valid files are found
    if total_runs > 0:
        success_rate = (success_count / total_runs) * 100
    else:
        success_rate = 0.0

    print("-" * 30)
    print(f"Directory Scanned: {directory}")
    print(f"Total Runs Found:  {total_runs}")
    print(f"Successes:         {success_count}")
    print(f"Failures:          {failure_count}")
    print(f"Success Rate:      {success_rate:.2f}%")
    print("-" * 30)
```
